```python
import nltk
import numpy as np
from os import path
from nltk.corpus import inaugural
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
ps = PorterStemmer()
sentences=inaugural.sents(fileids='1961-Kennedy.txt') 
sentences=[" ".join(list_of_words) for list_of_words in sentences]
for sentence in sentences:
    print("Sentence in consideration\n")
    print(sentence)
    words = word_tokenize(sentence)
    print("POS tagged Sentence:\n")
    print(nltk.pos_tag(words))
    print("Words-stemwords pair\n")
    for w in words:
        print(w+" : "+ps.stem(w)+"\n")
    print("\n\n##########################################################################################\n\n\\n\n")
```
